Define a procedure called "check_input" that takes an input as a square list of lists (2D Matrix of N x N) and returns a boolen "true" if the input is valid or "false" otherwise, according to the following criteria.

1.   Each coloumn must contain numbers from between 1 and N, with no duplicates.
2.   Each row must contain numbers from between 1 and N, with no duplicates.

```python
def check_input(arr):
    cols = 0
    rows = 0
    arrlen = len(arr[0])

    for j in range(0,3):
        for i in range(1,3):
            if(arr[0][j]==arr[i][j] or type(arr[i][j])==str):
                cols = cols + 1

    for i in range(0,arrlen):
        for j in range(1,arrlen):
            if(arr[i][0]==arr[i][j] or type(arr[i][j])==str):
                rows = rows + 1
    
    if(rows==0 and cols==0):
        return True
    else:
        return False





    
```

```python
# run this block to test code after defining the procedure

test1 = [[1, 2, 3],
        [2, 3, 1],
        [3, 1, 2]]

test2 = [[1, 2, 3, 4],
        [2, 3, 1, 3],
        [3, 1, 2, 3],
        [4, 4, 4, 4]]

test3 = [['a', 'b', 'c'],
        ['b', 'c', 'a'],
        ['c', 'a', 'b']]

print(check_input(test1)) # must return true
print(check_input(test2)) # must return false
print(check_input(test3)) # must return false
```

**Part A**- From the DataFrame created in Question 2, calculate the average marks for each subject

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/drive/My Drive/StudentMarkSheet.csv")
data=data.dropna(how='any')

AverageinBio = data[data['Exam name']=="Biology"]
print(AverageinBio)
AverageinBio=AverageinBio["Marks"]
AverageinBio=(AverageinBio.sum())/(len(AverageinBio))
print("Average in Biology = ",AverageinBio)

AverageinChemistry = data[data['Exam name']=="Chemistry"]
print(AverageinChemistry)
AverageinChemistry=AverageinChemistry["Marks"]
AverageinChemistry=(AverageinChemistry.sum())/(len(AverageinChemistry))
print("Average in Chemistry = ",AverageinChemistry)

AverageinMath = data[data['Exam name']=="Mathematics"]
print(AverageinMath)
AverageinMath=AverageinMath["Marks"]
AverageinMath=(AverageinMath.sum())/(len(AverageinMath))
print("Average in Mathematics = ",AverageinMath)

AverageinPhilosophy = data[data['Exam name']=="Philosophy"]
print(AverageinPhilosophy)
AverageinPhilosophy=AverageinPhilosophy["Marks"]
AverageinPhilosophy=(AverageinPhilosophy.sum())/(len(AverageinPhilosophy))
print("Average in Philosophy = ",AverageinPhilosophy)

AverageinSociology = data[data['Exam name']=="Sociology"]
print(AverageinSociology)
AverageinSociology=AverageinSociology["Marks"]
AverageinSociology=(AverageinSociology.sum())/(len(AverageinSociology))
print("Average in Sociology = ",AverageinSociology)

AverageinPhysics = data[data['Exam name']=="Physics"]
print(AverageinPhysics)
AverageinPhysics=AverageinPhysics["Marks"]
AverageinPhysics=(AverageinPhysics.sum())/(len(AverageinPhysics))
print("Average in Physics = ",AverageinPhysics)
```

**Part B** - Using the averages you just calculated, plot a Bar Graph showing the average marks of each subject

```python
courses=["Biology","Math","Chemistry","Sociology","Philosophy","Physics"]
avg = [AverageinBio,AverageinMath,AverageinChemistry,AverageinSociology,AverageinPhilosophy,AverageinPhysics]
plt.bar(courses,avg,fc="lightgreen")
plt.title("Courses & Average")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show
```
